```python
def clean_tweet(tweet):
    tweet = re.sub(r"@[A-Za-z0-9]+",' ', tweet)
    tweet = re.sub(r"https?://[A-Za-z0-9./]+",' ', tweet)
    tweet = re.sub("RT ", " ", tweet)
    tweet = re.sub("\n", " ", tweet)
    tweet = re.sub("&amp;", "", tweet)
    tweet = re.sub("_", "", tweet)
    tweet = re.sub("â€™t", "", tweet)
    tweet = tweet.replace(":", "")
    tweet = tweet.replace(",", "")
    tweet = tweet.replace("!", "")
    tweet = tweet.replace("#", "")
    tweet = tweet.replace(' ’ ', "'")
    tweet = tweet.replace(" ' ", "")
    tweet = tweet.replace(".", "")
    tweet = tweet.replace(",", "")
    tweet = tweet.replace("?", "")
    tweet=tweet.lower()
    return tweet
```
